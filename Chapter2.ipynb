Boston housing data
-------------------

```python
housing = pd.read_csv('Boston.csv')
```

```python
housing.describe()
```

```python
housing.columns.values
```

```python
housing.head()
```

Now let's drop that first column as it's a unnecessary one, which got imported accidentaly from the csv file.

```python
housing_clean = housing.drop('Unnamed: 0', axis=1 )
```

We have 506 observations and 14 variables.

1. CRIM: per capita crime rate by town 
2. ZN: proportion of residential land zoned for lots over 25,000 sq.ft. 
3. INDUS: proportion of non-retail business acres per town 
4. CHAS: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise) 
5. NOX: nitric oxides concentration (parts per 10 million) 
6. RM: average number of rooms per dwelling 
7. AGE: proportion of owner-occupied units built prior to 1940 
8. DIS: weighted distances to five Boston employment centres 
9. RAD: index of accessibility to radial highways 
10. TAX: full-value property-tax rate per 10,000 dollars 
11. PTRATIO pupil-teacher ratio by town 
12. B 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town 
13. LSTAT  lower status of the population 
14. MEDV Median value of owner-occupied homes in $1000's

```python
scatter_matrix(housing_clean, alpha=0.2, figsize=(16,16), diagonal='hist')
```

Exploring criminality: with lower median value of the house, the crime rate increases, also when the population has a big fraction of lower status population, crime rate increases. Criminality is higher in older houses and also is increasing when the distance to employment center is small. Dirtier areas seem to attract criminality (nox).

```python
housing_clean["chas"].value_counts() # 1 are bound by the charles river
```

```python
housing_clean["crim"].hist(bins=25)
```

```python
high_crime = housing_clean.loc[housing_clean['crim'] > 0.2]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
high_crime["ptratio"].hist(bins = 20, ax=ax, alpha = 0.5)
housing["ptratio"].hist(bins = 20, ax=ax, color='blue', alpha=0.3)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
high_crime["tax"].hist(bins = 25, ax=ax, alpha = 0.5)
housing["tax"].hist(bins = 25, ax=ax, color='blue', alpha=0.3)
```
